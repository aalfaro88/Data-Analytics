```python
import pandas as pd
data=pd.read_csv("marketing_customer_analysis.csv")
import time
from datetime import date

def clean_data(data):
    nan_data=data.isna().sum()/len(data)*100 
    nan_data=nan_data.reset_index()
    nan_data.columns=("CATEGORY","NANs")
    columns_drop=nan_data[nan_data["NANs"]>20]["CATEGORY"]
    data=data.drop(columns_drop,axis=1)
    
    data=data.rename(columns={"EmploymentStatus":"EMPLOYMENT STATUS"})
    data=data.drop(["Unnamed: 0"],axis=1)
    data.columns=data.columns.str.upper()
    data["STATE"]=data["STATE"].fillna("Unknown")
    data["RESPONSE"]=data["RESPONSE"].fillna("No Info")
    data["MONTHS SINCE LAST CLAIM"]=data["MONTHS SINCE LAST CLAIM"].fillna("No Claims")
    data["NUMBER OF OPEN COMPLAINTS"]=data["NUMBER OF OPEN COMPLAINTS"].fillna(0)
    data["VEHICLE CLASS"]=data["VEHICLE CLASS"].fillna("NO INFO")
    data["VEHICLE SIZE"]=data["VEHICLE SIZE"].fillna("NO INFO")
    data["EFFECTIVE TO DATE"]=pd.to_datetime(data["EFFECTIVE TO DATE"],errors="coerce")
    data["MONTH"]=data["EFFECTIVE TO DATE"].dt.month
    
    return data
    
```

```python
clean_data=clean_data(data)
```
