```python
#Create a Python connection with SQL database and retrieve the results of the last two queries (also mentioned below) as dataframes:

#Check the number of rentals for each customer for May

#Check the number of rentals for each customer for June
#Hint: You can store the results from the two queries in two separate dataframes.
```

```python
query='SELECT customer_id, count(rental_id) AS "rentals_may" \
FROM rentals_may \
GROUP BY customer_id \
ORDER BY customer_id ASC;'

rentals_may=pd.read_sql_query(query,engine)
rentals_may
```

```python
query='SELECT customer_id, count(rental_id) AS "rentals_june" \
FROM rentals_june \
GROUP BY customer_id \
ORDER BY customer_id ASC;'

rentals_june=pd.read_sql_query(query,engine)
rentals_june
```

```python
#Write a function that checks if customer borrowed more or less films in the month of 
#June as compared to May.

#Hint: For this part, you can create a join between the two dataframes created before, 
#using the merge function available for pandas dataframes. 
```

```python
rentals_comparison = pd.merge(rentals_may, rentals_june, on='customer_id')
rentals_comparison
```

```python
rentals_comparison["difference"]=rentals_comparison["rentals_june"]-rentals_comparison["rentals_may"]   
```

```python
rentals_comparison
```

```python
def categorydifference(x):
    if x > 0:
        return "+"
    elif x < 0:
        return "-"
    else:
        return "="
    
rentals_comparison["difference"]=list(map(categorydifference,rentals_comparison["difference"]))
rentals_comparison
```
