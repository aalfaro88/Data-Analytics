```python
artist_names = []
offset = 0
limit = 50
genres = ['pop', 'rock', 'hip-hop', 'country', 'jazz', 'classical', 'electronic', 'reggae', 'r&b', 'folk', 'blues', 'metal', 'punk', 'soul', 'funk']

for genre in genres:
    counter = 0
    while counter < 50:
        results = sp.search(q='genre:' + genre, type='artist', limit=limit, offset=offset)
        artists = results['artists']['items']
        for artist in artists:
            artist_names.append(artist['name'])
        offset += limit
        counter += len(artists)
    offset = 0
    print(genre,"done")
print(artist_names)
```
